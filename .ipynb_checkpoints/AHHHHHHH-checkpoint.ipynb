# Discussion of Data

The data used in this project was found on kaggle. The downloaded CSV file contained 65,574 team box scores, from the 1996-97 season, all the way to 2023-24. This dataset contained the boxscore stats from each individual game, distinguished by gameid. There were 29 total columns in this original dataset, including home, points, rebounds, assists, and all other basic counting stats.  

There were two rows per game, they were matched on game_id, and date, but had different values for each of the columns, because they were from the perspecitve of whichever team.  

I took care of all of essentially all of my data preperation using R, more specifically, a package called tidyverse, which enables SQL like queries. I After I had loaded the dataset in, I created the target variable, which is result, I labeled it temporarily as a W if teamA and more points than teamB for the same given gameID. I created two additional variables, called back_to_back and win%. Back_to_back was a dummy variable, it grouped team and season, and checked to see if a team played a game in 2 consecutive games, where the second game would be a 1, and all other cases, including the first game would be a 0. Win% merely keeps track of the win% of a given game, it seems a lot like result, but I would need it for when I figured out season stats.  

After I had organized the plain boxscores dataset, I then created a new data frame, "season_stats", which took the means of each column, for each team, for each season.  

The objective for this project was to take recent statistics, to predict the outcome of the next game. When we are past the 10th game of the season, this was easy, all what you have to do is take the mean of the last 10 games for each statsitic. When it is the first game of the season however, this is a little difficult. There are no previous games for a given season. I was left with a few options of how to deal with this problem, the first being to just truncate all of the game 1s for a season, but I didn't want to do this. Now although a lot changes from season to season for an NBA team, I thought a more logical idea was to include the previous season statistics for just the first game.  

For this logic, I implemented a method, that grouped the dataframe by team and season, essentially going by team, by season, it would repetetively fill in values in an empty dataframe. For each of the columns, if the game number for a given season was equal to 1, it would use the previous years stats (averages), and if the game number was = x and x was <= than 10, it would use the previous x-1 game averages. If game number was = y, and y > 10, it would use the previous 10 games averages.  

After this method ran, and the new dataset was created, I then added in some different percentages: TOV%, ORB%, DRB%, FT%, 3P%, eFG%, and then Defensive Rating, Offensive rating, and net rating. Then I joined all of the boxscores, on game_id, with whatever team being listed first haivng their columns renamed as {column}.x, and the seconary team having {column}.y.  

This complete dataset has the averages for each team, their opponents averages (over x games), and their opponent for a given game averages, and their opponents, opponents (averages over y games)  

I only included games after the 2015 season, to avoid major differences in style of play (main example is amount of 3s each team shoots over the last 10 years. The complete datset had 11,979 total regular season box scores, and 28 total columns.  

https://www.kaggle.com/datasets/szymonjwiak/nba-traditional

### Relationship of Back to Back Games and Games Won

```python
# Make a clean copy
data_copy2 = data.copy()

# Create 'result.y' if not already created
data_copy2['result.y'] = 1 - data_copy2['result']  # Or your correct logic

# Combine both teams' data
combined_b2b = pd.concat([data_copy2['back_to_back.x'], data_copy2['back_to_back.y']], ignore_index=True)
combined_result = pd.concat([data_copy2['result'], data_copy2['result.y']], ignore_index=True)

# Create contingency table from combined data
btb_info_combined = pd.crosstab(combined_b2b, combined_result)

# Rename index and columns
btb_info_combined.index = ["Not Back-to-Back", "Back-to-Back"]
btb_info_combined.columns = ["Losses", "Wins"]

# Add totals
btb_info_combined["Total Games"] = btb_info_combined["Losses"] + btb_info_combined["Wins"]
btb_info_combined.loc["Total"] = btb_info_combined.sum(numeric_only=True)
btb_info_combined.loc["Total", "Total Games"] = btb_info_combined.loc["Total", ["Losses", "Wins"]].sum()

# Optional: Add win rate
btb_info_combined["Win Rate (%)"] = (
    btb_info_combined["Wins"] / btb_info_combined["Total Games"] * 100
).round(2)
btb_info_combined.loc["Total", "Win Rate (%)"] = ""

# Display
print("Combined Game Results by Back-to-Back Status (Both Teams Included)\n")
print(btb_info_combined)
```

### Relationship of Home Games and Games Won

```python
data_copy2 = data.copy()
data_copy2['result.y'] = 1 - data_copy2['result']
data_copy2['home.y'] = 1 - data_copy2['home.x']  # Create home.y

# Combine both teams' home status and results
combined_home = pd.concat([data_copy2['home.x'], data_copy2['home.y']], ignore_index=True)
combined_result2 = pd.concat([data_copy2['result'], data_copy2['result.y']], ignore_index=True)

# Create contingency table
home_info_combined = pd.crosstab(combined_home, combined_result2)

# Rename index and columns
home_info_combined.index = ["Not Home", "Home"]
home_info_combined.columns = ["Losses", "Wins"]

# Add totals
home_info_combined["Total Games"] = home_info_combined["Losses"] + home_info_combined["Wins"]
home_info_combined.loc["Total"] = home_info_combined.sum(numeric_only=True)
home_info_combined.loc["Total", "Total Games"] = home_info_combined.loc["Total", ["Losses", "Wins"]].sum()

# Add win rate
home_info_combined["Win Rate (%)"] = (
    home_info_combined["Wins"] / home_info_combined["Total Games"] * 100
).round(2)
home_info_combined.loc["Total", "Win Rate (%)"] = ""

# Display
print("Combined Game Results by Home Status (Both Teams Included)\n")
print(home_info_combined)
```

# Preparing data for statistical modeling

### Remove games before 5th game of the season

```python
data = data[data["game_number.x"] > 5] # This is so there are at least 5 games of data to generate the dataset
data = data[data["season"] >= 2020] # This is so there are at least 5 games of data to generate the dataset
```

### Standardize variables

```python
from sklearn.preprocessing import StandardScaler

# Drop the 'result' column to get features
X = data.drop(columns=['result'])
# Define the target variable
y = data['result']

# Columns to exclude from scaling
exclude_cols = ['season','home.x', 'back_to_back.x', 'back_to_back.y', "game_number.x", 
                "game_number.y", "date", "team.x", "team.y"]

# Separate the columns to scale and those to leave as-is
X_to_scale = X.drop(columns=exclude_cols)
X_excluded = X[exclude_cols]

# Standardize only the desired features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(X_to_scale)

# Reconstruct the scaled DataFrame and concatenate with excluded columns
X_scaled_df = pd.DataFrame(scaled_features, columns=X_to_scale.columns, index=X.index)
X_standardized_df = pd.concat([X_scaled_df, X_excluded], axis=1)

# Optional: reorder columns to match original X
X_standardized_df = X_standardized_df[X.columns]

X_standardized_df
```

### 2025 season data

```python
season_col = X_standardized_df['season']

train_mask = season_col < 2025
predict_mask = season_col == 2025

X_data = X_standardized_df[train_mask].drop(columns=['season'])
y_data = y[train_mask]

X_2025 = X_standardized_df[predict_mask].drop(columns=['season'])
y_2025 = y[predict_mask]


X_2025_full = X_2025  # your original full 2025 DataFrame
X_2025 = X_2025_full.drop(columns=['game_number.x', 'game_number.y', "date", "team.x", "team.y"])
```

### Train-Test-Split 70-30

Next, we create the training and testing datasets using an 80/20 split. The target variable, result, is separated from the feature dataset X and stored in y. 

```python
from sklearn.model_selection import train_test_split

# Train-test split (80-20)
X_train_full, X_test_full, y_train, y_test = train_test_split(X_standardized_df, y, test_size=0.2, random_state=42)

X_train = X_train_full.drop(columns=['game_number.x', 'game_number.y', "season", "date", "team.x", "team.y"])
X_test = X_test_full.drop(columns=['game_number.x', 'game_number.y', "season", "date", "team.x", "team.y"])
```

# Statistical modeling

## a) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV


# Create the logistic regression model
model = LogisticRegression(random_state=0)

param_grid = [
    {'penalty': ['l2'], 'C': [0.01, 0.1, 1, 10, 100], 'solver': ['lbfgs', 'newton-cg', 'sag', 'newton-cholesky']},
    {'penalty': ['l1'], 'C': [0.01, 0.1, 1, 10, 100], 'solver': ['liblinear', 'saga']},
    {'penalty': ['l2'], 'C': [0.01, 0.1, 1, 10, 100], 'solver': ['liblinear', 'saga']},
    {'penalty': ['elasticnet'], 'C': [0.01, 0.1, 1, 10, 100], 'solver': ['saga'], 'l1_ratio': [0.1, 0.5, 0.9]}
]

clf = GridSearchCV(model, param_grid, cv=5, n_jobs=-1, scoring='accuracy')
    
# Fit the model on the train set
clf.fit(X_train, y_train)

# Find the predictions for the test set and 2025
preds_log = clf.predict(X_test)
preds_log_2025 = clf.predict(X_2025)

print("Best Parameters:", clf.best_params_)
```

### Metrics and plots: Logistic Regression

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Start by finding the accuracy on the test set and then for the 2025 season
accuracy_log = accuracy_score(y_test, preds_log)
accuracy_log_2025 = accuracy_score(y_2025, preds_log_2025)

print(f"The accuracy of the logistic regression model on the test set was: {round(accuracy_log,3)}")
print(f"The accuracy of the logistic regression model on the 2025 season was: {accuracy_log_2025}")




# ADD MORE METRICS/PLOTS
```

```python
# Get best estimator from GridSearchCV
best_log_model = clf.best_estimator_

# Get the coefficients
coefficients = best_log_model.coef_[0]  # shape: (n_features,)

# Combine with feature names
feature_names = X_train.columns  # or whatever DataFrame you used to train
coef_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients
})

# Sort by absolute value of coefficient to see most influential features
coef_df['Abs_Coefficient'] = coef_df['Coefficient'].abs()
coef_df_sorted = coef_df.sort_values(by='Abs_Coefficient', ascending=False)

# Display top features
print(coef_df_sorted)
```

## b) SVM classifier

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold

# Initialize SVM model
svc_model = SVC(random_state=42)

# Define a simpler parameter grid for SVM
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree':[3,4,6,7,9],
    'gamma': ['scale', 0.01, 0.1],
}

# Run the grid search
svc_clf = GridSearchCV(svc_model, param_grid, cv=5, n_jobs=-1, scoring='accuracy')

# Fit the model with the best paramaters
svc_clf.fit(X_train, y_train)

# Find the predictions for the test set and 2025
preds_SCV = svc_clf.predict(X_test)
preds_SVC_2025 = svc_clf.predict(X_2025)

# Best model cross validation score
print("Best Parameters:", svc_clf.best_params_)
```

### Metrics and plots: SVC

```python
# Start by finding the accuracy on the test set and then for the 2025 season
accuracy_SVC = accuracy_score(y_test, preds_SCV)
accuracy_SVC_2025 = accuracy_score(y_2025, preds_SVC_2025)

print(f"The accuracy of the SVM model on the test set was: {round(accuracy_SVC,3)}")
print(f"The accuracy of the SVM model on the 2025 season was: {accuracy_SVC_2025}")





# ADD MORE METRICS / PLOTS / FEATURE IMPORTANCE
```

## c) Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Define Random Forest model
rf = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_features': ['sqrt', 'log2'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_depth': [None, 10, 20],
    'class_weight': [None, 'balanced']
}

# Perform Grid Search with Cross-Validation
rf_clf = GridSearchCV(rf, param_grid, cv=5, n_jobs=-1)

# Fit the model with the best paramaters
rf_clf.fit(X_train, y_train)

# Find the predictions for the test set and 2025
preds_RF = rf_clf.predict(X_test)
preds_RF_2025 = rf_clf.predict(X_2025)

# Best model cross validation score
print("Best Parameters:", rf_clf.best_params_)
```

```python
# Start by finding the accuracy on the test set and then for the 2025 season
accuracy_RF = accuracy_score(y_test, preds_RF)
accuracy_RF_2025 = accuracy_score(y_2025, preds_RF_2025)

print(f"The accuracy of the Random Forest model on the test set was: {round(accuracy_RF,3)}")
print(f"The accuracy of the Random Forest model on the 2025 season was: {accuracy_RF_2025}")


# ADD MORE METRICS / PLOTS / FEATURE IMPORTANCE
```

## XGBoost

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV

# Initialize model
XGB_model = XGBClassifier(eval_metric='logloss')

# Build paramater grid
param_grid = {
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 4, 5, 6],
    "learning_rate": [0.01, 0.05, 0.1],
    "reg_alpha": [0, 0.1, 1],              # L1 regularization
    "reg_lambda": [0, 0.1, 1],             # L2 regularization
    "subsample": [0.7, 0.8, 1.0],          # row sampling
    "colsample_bytree": [0.7, 0.8, 1.0],   # feature sampling
    "gamma": [0, 0.1, 0.5],                # minimum loss reduction to split
}


# Perform Grid Search with Cross-Validation
XGB_clf = RandomizedSearchCV(XGB_model, param_grid, cv=5, n_jobs=-1)

# Fit the model with the best paramaters
XGB_clf.fit(X_train, y_train)

# Find the predictions for the test set and 2025
preds_XGB = XGB_clf.predict(X_test)
preds_XGB_2025 = XGB_clf.predict(X_2025)

# Best model cross validation score
print("Best Parameters:", XGB_clf.best_params_)
```
