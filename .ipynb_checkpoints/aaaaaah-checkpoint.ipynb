```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_grouped_histogram(df, metric_col_x, metric_col_y, result_col="result", bins=30, shrink=0.8,
                            x_label=None, title=None):
    """
    Plots a grouped histogram of a metric by result outcome using both .x and .y team data.
    """
    metric_values = pd.concat([df[metric_col_x], df[metric_col_y]], ignore_index=True)
    results = pd.concat([df[result_col], df[f"{result_col}.y"]], ignore_index=True)
    teams = [".x"] * len(df) + [".y"] * len(df)

    plot_df = pd.DataFrame({
        "Metric": metric_values,
        "Result": results.map({1: "Win", 0: "Loss"}),
        "Team": teams
    })

    # Use default label if none provided
    metric_label = x_label or metric_col_x[:-2]

    plt.figure(figsize=(10, 6))
    sns.histplot(data=plot_df, x="Metric", hue="Result", multiple="dodge", bins=bins, 
                 palette={"Win": "blue", "Loss": "red"}, shrink=shrink)
    plt.title(title or f"Grouped Histogram of {metric_label} by Game Result")
    plt.xlabel(metric_label)
    plt.ylabel("Game Count")
    plt.grid(True)
    plt.show()
```

## b. Scatterplots

```python
# Plots two scatter plots side by side comparing a team's metric vs. opponent's metric for wins and losses.
def plot_metric_comparison(df, metric_name, result_col_1="result", result_col_2="result.y",
                           suffix_1=".x", suffix_2=".y", figsize=(10, 5)):
    
    team_metric_1 = f"{metric_name}{suffix_1}"
    opp_metric_1 = f"opp_{metric_name}{suffix_1}"
    team_metric_2 = f"{metric_name}{suffix_2}"
    opp_metric_2 = f"opp_{metric_name}{suffix_2}"
    
    fig, axes = plt.subplots(1, 2, figsize=figsize)

    # Team 1 plot
    axes[0].scatter(df[df[result_col_1] == 1][team_metric_1], df[df[result_col_1] == 1][opp_metric_1],
                    marker=".", s=75, c="blue", label="Win")

    axes[0].scatter(df[df[result_col_1] == 0][team_metric_1], df[df[result_col_1] == 0][opp_metric_1],
                    marker=".", s=75, c="red", label="Loss")

    axes[0].set_title(f"{metric_name} of last 10 games for {suffix_1} and opponent")
    axes[0].set_xlabel("Team " + metric_name)
    axes[0].set_ylabel("Opponent " + metric_name)
    axes[0].grid()
    axes[0].legend()

    # Team 2 plot
    axes[1].scatter(df[df[result_col_2] == 1][team_metric_2], df[df[result_col_2] == 1][opp_metric_2],
                    marker=".", s=75, c="blue", label="Win")

    axes[1].scatter(df[df[result_col_2] == 0][team_metric_2], df[df[result_col_2] == 0][opp_metric_2],
                    marker=".", s=75, c="red", label="Loss")

    axes[1].set_title(f"{metric_name} of last 10 games for {suffix_2} and opponent")
    axes[1].set_xlabel("Team " + metric_name)
    axes[1].set_ylabel("Opponent " + metric_name)
    axes[1].grid()
    axes[1].legend()
    plt.tight_layout()
    plt.show()
```

## c. Tables

```python
def summarize_game_outcomes(data, status_col_base, result_col="result", status_labels=None, title=""):
    """
    Summarizes game outcomes (Wins/Losses) by a binary status (e.g., back_to_back, home).
    
    Parameters:
    - data: Original DataFrame with .x and .y columns
    - status_col_base: The base name of the status column (e.g., "back_to_back", "home")
    - result_col: The name of the result column (default: "result")
    - status_labels: Optional list of labels for index (e.g., ["Not B2B", "B2B"])
    - title: Optional title to print
    
    Returns:
    - Formatted DataFrame with Wins, Losses, Total Games, and Win Rate (%)
    """
    # Make a copy
    df = data.copy()
    df[f"{result_col}.y"] = 1 - df[result_col]  # ensure .y version of result exists
    
    # Create home.y
    df['home.y'] = 1 - df['home.x']  # Create home.y

    # Combine status and result columns for both teams
    combined_status = pd.concat([df[f"{status_col_base}.x"], df[f"{status_col_base}.y"]], ignore_index=True)
    combined_result = pd.concat([df[result_col], df[f"{result_col}.y"]], ignore_index=True)

    # Create contingency table
    table = pd.crosstab(combined_status, combined_result)

    # Rename rows and columns
    table.index = status_labels or ["Status 0", "Status 1"]
    table.columns = ["Losses", "Wins"]

    # Add totals and win rate
    table["Total Games"] = table["Wins"] + table["Losses"]
    table.loc["Total"] = table.sum(numeric_only=True)
    table.loc["Total", "Total Games"] = table.loc["Total", ["Losses", "Wins"]].sum()
    table["Win Rate (%)"] = (table["Wins"] / table["Total Games"] * 100).round(2)
    table.loc["Total", "Win Rate (%)"] = np.nan

    # Display
    if title:
        print(f"\n{title}\n")
    print(table)

    return table
```

# 3. Data Preperation 

### Filter Data

```python
# Remove games before 5th game of the season, and only focus on games after the 2020 season
data = data[data["game_number.x"] > 5]
data = data[data["season"] >= 2020]
```

### Remove features with very high correlation

```python
# Net rating, win%, and elo are all highly correlated, elo more accurately trends a teams success during a season
data = data.drop(columns=['win%.x','win%.y'])
```

### Prepare Model Data and 2025 Data

```python
# Initialize X (data) and y (target variable)
X = data.drop(columns=['result'])
y = data['result'] 

# Specify which columns to not include in model creation
exclude_cols = ['season', "date", "game_number.x", "team.x", 'home.x','back_to_back.x', 
                "game_number.y", "team.y", 'back_to_back.y']

# We want to make two seperate datasets, one that the model will be trained on (before 2025), and 2025
# Prepare for scaling and drop cols
X_model_unscaled = X[X["season"] < 2025].drop(columns=exclude_cols)
X_2025_unscaled = X[X["season"] == 2025].drop(columns=exclude_cols)

X_model_excluded = X[X["season"] < 2025][exclude_cols]
X_2025_excluded = X[X["season"] == 2025][exclude_cols]

y_model = y[X["season"] < 2025]
y_2025 = y[X["season"] == 2025]
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit scaler on model data only
scaler = StandardScaler()

# Standardize the model data 
X_model_scaled = pd.DataFrame(scaler.fit_transform(X_model_unscaled), 
                              columns=X_model_unscaled.columns, 
                              index=X_model_unscaled.index)

# 2025 data transformed on the scaled data
X_2025_scaled = pd.DataFrame(scaler.transform(X_2025_unscaled), 
                             columns=X_2025_unscaled.columns, 
                             index=X_2025_unscaled.index)


# Doing it this way avoids leaky data. The data from the 2020-24 seasons are standardized, and then 2025 is fitted?
```

### Make 2025 dataset that contain all of the columns (For Analysis)

```python
# These will be used later, when we want to analyze the results
X_FULL = pd.concat([X_model_excluded, X_model_scaled], axis=1)
X_2025_FULL = pd.concat([X_2025_excluded, X_2025_scaled], axis=1)

# Add back in the target variable
X_FULL["result.x"] = y_model.values  # Or y_2025.reset_index(drop=True)
X_2025_FULL["result.x"] = y_2025.values  # Or y_2025.reset_index(drop=True)
```

### Train-Test-Split 70-30

```python
from sklearn.model_selection import train_test_split

X_train_FULL, X_test_FULL, y_train, y_test = train_test_split(
    X_FULL, y_model, test_size=0.2, random_state=42
)

# Drop all of the columns that we don't want to use for modeling
cols_to_drop = ["game_number.x", "game_number.y", "season", "date", "team.x", "team.y", "result.x"]

# These are the final datasets to be used in predicting
X_train = X_train_FULL.drop(columns=cols_to_drop)
X_test = X_test_FULL.drop(columns=cols_to_drop)
X_2025 = X_2025_FULL.drop(columns=cols_to_drop)
```

# 4. Statistical modeling

## a) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV, RandomizedSearchCV

# Create TimeSeriesSplit object
tscv = TimeSeriesSplit(n_splits=5)

# Create the logistic regression model
pipe_logreg = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(random_state=0))
])

param_grid_logreg = [
    {'logreg__penalty': ['l2'], 'logreg__C': [0.01, 0.1, 1, 10, 100], 'logreg__solver': ['lbfgs', 'newton-cg', 'sag', 'newton-cholesky'], 'logreg__class_weight': [None, 'balanced']},
    {'logreg__penalty': ['l1'], 'logreg__C': [0.01, 0.1, 1, 10, 100], 'logreg__solver': ['liblinear', 'saga'], 'logreg__class_weight': [None, 'balanced']},
    {'logreg__penalty': ['l2'], 'logreg__C': [0.01, 0.1, 1, 10, 100], 'logreg__solver': ['liblinear', 'saga'], 'logreg__class_weight': [None, 'balanced']},
    {'logreg__penalty': ['elasticnet'], 'logreg__C': [0.01, 0.1, 1, 10, 100], 'logreg__solver': ['saga'], 'logreg__l1_ratio': [0.0, 0.1, 0.25, 0.5, 0.75, 0.9, 1.0], 'logreg__class_weight': [None, 'balanced']}
]

logreg_clf = GridSearchCV(pipe_logreg, param_grid_logreg, cv=tscv, n_jobs=-1, scoring='accuracy')
    
# Fit the model on the train set
logreg_clf.fit(X_train, y_train)

print("Best Parameters:", logreg_clf.best_params_)
```

### Metrics and plots

```python
# Import metrics
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix,
                             precision_score, recall_score, f1_score, log_loss)
```

```python
def evaluate_model(clf, X_test, y_test, X_2025, y_2025):
    # Evaluate a trained model on test and 2025 datasets
    
    # Predict labels and probabilities
    preds_clf = clf.predict(X_test)
    preds_clf_2025 = clf.predict(X_2025)
    probs_clf = clf.predict_proba(X_test)
    probs_clf_2025 = clf.predict_proba(X_2025)

    print(f"The accuracy of the model on the test set was: {accuracy_score(y_test, preds_clf):.3f}")
    print(f"The accuracy of the model on the 2025 season was: {accuracy_score(y_2025, preds_clf_2025):.3f}")

    # --- Test Set Metrics ---
    print("\n===== Test Set Metrics =====")
    print(f"Accuracy: {accuracy_score(y_test, preds_clf):.3f}")
    print(f"Precision: {precision_score(y_test, preds_clf):.3f}")
    print(f"Recall: {recall_score(y_test, preds_clf):.3f}")
    print(f"F1 Score: {f1_score(y_test, preds_clf):.3f}")
    print(f"Log Loss: {log_loss(y_test, probs_clf):.3f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, preds_clf))
    print("\nClassification Report:")
    print(classification_report(y_test, preds_clf))

    # --- 2025 Season Metrics ---
    print("\n===== 2025 Season Metrics =====")
    print(f"Accuracy: {accuracy_score(y_2025, preds_clf_2025):.3f}")
    print(f"Precision: {precision_score(y_2025, preds_clf_2025):.3f}")
    print(f"Recall: {recall_score(y_2025, preds_clf_2025):.3f}")
    print(f"F1 Score: {f1_score(y_2025, preds_clf_2025):.3f}")
    print(f"Log Loss: {log_loss(y_2025, probs_clf_2025):.3f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_2025, preds_clf_2025))
    print("\nClassification Report:")
    print(classification_report(y_2025, preds_clf_2025))

    return preds_clf, preds_clf_2025, probs_clf, probs_clf_2025
```

```python
preds_log, preds_log_2025, probs_log, probs_log_2025 = evaluate_model(logreg_clf, X_test, y_test, X_2025, y_2025)
```

**TP = True Positives (predicts a win, is a win)  
TN = True Negatives (predicts a loss, is a loss)  
FP = False positives (Predicts a win, is a loss)  
FN = False Negatives (Predicts a loss, is a win)**  
  
**Accuracy**  
- (TP + TN) / (TP + TN + FP + FN)  
- Correct predictions / all predictions  
- The proportion of total predictions the model got right.  

**Precision**  
- TP / (TP + FP)
- Of the predictions labeled positive (e.g. wins), how many were actually correct.  

**Recall**  
- TP / (TP + FN)
- Of all the actual positives (e.g. true wins), how many did the model correctly identify.  

**F1 Score**  
- 2* (Precision * recall) / (Precision + Recall)
- The harmonic mean of precision and recall  balances the two when both matter.

**Log Loss**
- How well the model's probability estimates match the actual outcomes  lower is better.

# How many times was the model correct for each team

```python
teams = ["ATL", "BOS", "BRK", "CHO", "CHI", "CLE", 
         "DAL", "DEN", "DET", "GSW", "HOU", "IND", 
         "LAC", "LAL", "MEM", "MIA", "MIL", "MIN",
         "NOP", "NYK", "OKC", "ORL", "PHI", "PHO", 
         "POR", "SAC", "SAS", "TOR", "UTA", "WAS"]

team_stats = []

for team in teams:
    # Boolean masks for appearances
    mask_x = data_2025["team.x"] == team
    mask_y = data_2025["team.y"] == team

    # Game count
    team_game_count = (mask_x | mask_y).sum()

    # Win count: .x wins when result.x == 1, .y wins when result.x == 0
    team_win_count = ((mask_x & (data_2025["result.x"] == 1)) |
                      (mask_y & (data_2025["result.x"] == 0))).sum()

    # Logistic model correct prediction
    log_correct = ((mask_x & (data_2025["log_pred"] == data_2025["result.x"])) |
                   (mask_y & (data_2025["log_pred"] != data_2025["result.x"]))).sum()
    log_accuracy = log_correct / team_game_count

    # SVC model correct prediction
    svc_correct = ((mask_x & (data_2025["SVC_pred"] == data_2025["result.x"])) |
                   (mask_y & (data_2025["SVC_pred"] != data_2025["result.x"]))).sum()
    svc_accuracy = svc_correct / team_game_count

    # Append result
    team_stats.append({
        "team": team,
        "game_count": team_game_count,
        "games_won": team_win_count,
        "log_accuracy": round(log_accuracy, 4),
        "svc_accuracy": round(svc_accuracy, 4)
    })

# Create final DataFrame
df = pd.DataFrame(team_stats)

# Sort by wins or accuracy
df = df.sort_values(by="games_won", ascending=False).reset_index(drop=True)

# Display result
print(df)
```
