```python
import pandas as pd
import time
import random
```

```python
def pull_reg_data(team, year):
    url = f"https://www.basketball-reference.com/teams/{team}/{year}/gamelog/"
    try:
        data = pd.read_html(url)
        reg_szn = data[0]
        
       # reg_szn = reg_szn.drop(index=reg_rows_to_drop)
        return reg_szn
    
    except Exception as e:
        print(f"Error retrieving data for {team} in {year}: {e}")
        return pd.DataFrame()  # Return an empty DataFrame in case of error
```

```python
boxscores_complete_new = pd.DataFrame() # 2020-2024
```

```python
teams = ["ATL", "BOS", "BRK", "CHO", "CHI", "CLE", 
         "DAL", "DEN", "DET", "GSW", "HOU", "IND", 
         "LAC", "LAL", "MEM", "MIA", "MIL", "MIN",
         "NOP", "NYK", "OKC", "ORL", "PHI", "PHO", 
         "POR", "SAC", "SAS", "TOR", "UTA", "WAS"]

years = list(range(2020, 2026))  # RUN THIS AGAIN OFFICIALLY

for year in years:
    random.shuffle(teams)  # Shuffle team order for the current year

    for team in teams:
        attempts = 0
        success = False

        while attempts < 3 and not success:
            temp = pull_reg_data(team, year)  # Replace with your data retrieval function
            temp["team"] = team
            temp["season"] = year

            if not temp.empty:
                boxscores_complete_new = pd.concat([boxscores_complete_new, temp], ignore_index=True)
                success = True
                print(f"Data retrieved successfully for {team} in {year}.")
            else:
                attempts += 1
                print(f"Attempt {attempts} failed for {team} in {year}. Retrying in 10 seconds.")
                time.sleep(10)

        if not success:
            print(f"Failed to retrieve data for {team} in {year} after 3 attempts.")

        time.sleep(6)

    print(f"Completed data collection for {year}.")
```
